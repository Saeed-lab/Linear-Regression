# What is Matplotlib?

It is a multiplatform data visualization library.

```python
import matplotlib.pyplot as plt 
import numpy as np
#plt.style.use('classic') # Classic Style
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
plt.plot( x, np.exp(x), color = 'green')
```

```python
x= np.linspace(-4*np.pi, 4*np.pi, 10000 )

plt.plot(x, np.cos(x))
```

```python
fig = plt.figure()
x = np.linspace(-10, 10, 1000)

plt.plot( x, np.exp(x), color = 'green')




plt.xlabel("x-axis")
plt.ylabel("y-labe")

plt.ylim(-100, 2000)
plt.xlim(-9, 9)
```

```python
y = np.linspace(-10, 10, 50)
plt.scatter(y, y+2, label = "Line")
```

## Labeling

```python

plt.title("Sinusoidal functions")
plt.plot(x, np.cos(x), linestyle = '-', label = "Cos(x)")
plt.plot(x, np.sin(x), linestyle = '--')
plt.axis('equal')
plt.legend()
```

## Subplots

```python

plt.subplot(3,1,1)
plt.plot(x, np.sin(x))
plt.axis("equal")



plt.subplot(3,1,2)
plt.plot(x, np.cos(x))
plt.axis("equal")

plt.subplot(3,1,3)
plt.plot(x, np.cos(x))
plt.axis("equal")
```

```python
fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(x, x)
ax2.plot(x, -x)
```
