```python
path_in = "https://raw.githubusercontent.com/abdoghareeb46/Predict-Customer-Churn/master/ChurnData.csv"
```

```python
df = pd.read_csv(path_in)
```

```python
df = df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip',   'callcard', 'wireless','churn']]
#churn_df['churn'] =df['churn'].astype('int')
df.head()
```

```python
df.shape
```

```python
X = df.drop('churn', 1)
y = df.churn
```

```python
lg = LogisticRegression(C=0.01, solver='liblinear')
```

```python
lg.fit(X, y)
```

```python
yhat = lg.predict(X)
```

```python
yhat
```

```python
yhat_prob = lg.predict_proba(X)
yhat_prob
```

```python
confusion_matrix(y, yhat)
```
